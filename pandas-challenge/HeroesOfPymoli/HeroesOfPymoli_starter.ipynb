```python

import pandas as pd

file_to_load = "resources/purchase_data.csv"


game_df = pd.read_csv(file_to_load)
game_df.head()
```

## Player Count

* Display the total number of players


```python
player_total = len(game_df["SN"].value_counts())
player_total_df = pd.DataFrame(columns = ['Total Players'])
player_total_df.loc[0,'Total Players'] = player_total
player_total_df.style.set_properties(**{'text-align': 'center'})

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_count = len(game_df["Item Name"].value_counts()) 
avg_price = game_df['Price'].mean()
purchases = game_df['Price'].count()
revenue = game_df['Price'].sum()
summary_df = pd.DataFrame({"Number of Unique Items": item_count,
                              "Average Price": avg_price,
                              "Number of Purchases": purchases,
                              "Total Revenue": revenue}, index=[0])

summary_df



```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


game_3["Total Count"] = pd.cut(game_3["Age"], bins, labels=group_names, include_lowest=True)
Total_Count_df = pd.DataFrame(game_3["Total Count"].value_counts())
Total_Count_df["Percentage of Players"] = ((Total_Count_df["Total Count"]/(Total_Count_df["Total Count"].sum())) * 100).round(2)
Total_Count_df.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
group_2 = game_df.copy()

# Set binning and index values
group_2['Age']=pd.cut(game_df['Age'], bins, labels=group_names)

# Calculate totals 
group_3 = group_2[['Price', 'Age']].groupby(['Age']).count()
#rename the column
group_3.rename(columns={'Price': 'Purchase Count'},inplace = True)

#calculate average purchase price 
group_3["Average Purchase Price"] = group_2[['Price', 'Age']].groupby(['Age']).mean()

#calculate total purchase price 
group_3['Total Purchase Price'] = group_2[['Price', 'Age']].groupby(['Age']).sum()

group_3['Purchase Total per Person'] = group_3['Total Purchase Price'] / Total_Count_df['Total Count']



group_3['Average Purchase Price'] = group_3['Average Purchase Price'].map('${:,.2f}'.format)
group_3['Total Purchase Price'] = group_3['Total Purchase Price'].map('${:,.2f}'.format)
group_3['Purchase Total per Person'] = group_3['Purchase Total per Person'].map('${:,.2f}'.format)          


group_3

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
big_ballers_df = game_df['Item ID'].groupby(game_df['SN']).count()

# Set columns 
big_ballers_df = pd.DataFrame(data=big_ballers_df)
big_ballers_df.columns = ['Purchase Count']

# Calculate column values
big_ballers_df['Average Purchase Price'] = round(game_df['Price'].groupby(game_df['SN']).mean(),2)
#total purchase price for each screen name
big_ballers_df['Total Purchase Value'] = game_df['Price'].groupby(game_df['SN']).sum()

#sort dataframe in descending order
big_ballers_df.sort_values(by=['Total Purchase Value'], ascending=False, inplace=True)

# Formatting total price and average price
big_ballers_df['Average Purchase Price'] = big_ballers_df['Average Purchase Price'].map('${:,.2f}'.format)
big_ballers_df['Total Purchase Value'] = big_ballers_df['Total Purchase Value'].map('${:,.2f}'.format)

big_ballers_df.head()





```

```python
majority_df = game_df.groupby(['Item ID', 'Item Name','Price'])['Price'].agg(['count','sum'])

#results in two columns
majority_df.columns = ['Purchase Count', 'Total Purchase Value']

# formatting index 
majority_df.reset_index(inplace=True)
majority_df.set_index(['Item ID','Item Name'] ,inplace=True)


majority_df = majority_df[['Purchase Count', 'Price', 'Total Purchase Value']]

# Sorting in descending order 
popular_df = majority_df.sort_values(by='Purchase Count', ascending=False)

# Format the price and total purchase value 
popular_df['Price'] = popular_df['Price'].map('${:,.2f}'.format)
popular_df['Total Purchase Value'] = popular_df['Total Purchase Value'].map('${:,.2f}'.format)


popular_df.head()
```
